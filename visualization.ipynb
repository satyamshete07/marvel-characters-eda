```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("marvel_cleaned_dataset.csv")

sns.set(style="darkgrid")
```

### 1. Univariate - Gender Count

```python

# 1. Count of Characters by Gender
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df, palette="Set2")
plt.title("Distribution of Gender")
plt.show()

print("Observation: Most characters are Male compared to Female.")

```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Role", data=df, palette="Set1")
plt.title("Role Distribution")
plt.show()

print("Observation: Heroes dominate compared to Villains and Supporting characters.")

```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Power_Level"], bins=10, kde=True, color="purple")
plt.title("Power Level Distribution")
plt.show()

print("Observation: Most characters have medium to high power levels.")
```

### 4. Pie Chart :-

```python
plt.figure(figsize=(6,6))
df["Team"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue","orange","green"])
plt.title("Team Distribution")
plt.ylabel("")
plt.show()

print("Observation: Avengers team has the highest representation.")
```

### 2. Bivariate Analysis

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Role", y="Power_Level", data=df, palette="coolwarm")
plt.title("Role vs Power Level")
plt.show()

print("Observation: Heroes generally have higher power levels than supporting characters.")
```

```python
# -----------------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="First_Appearance_Year", y="Power_Level", hue="Role", data=df)
plt.title("Year vs Power Level")
plt.show()

print("Observation: Newer characters tend to have slightly higher power levels.")
```

### 3. Multivariate Analysis


```python
plt.figure(figsize=(6,4))
corr = df[["Power_Level", "Popularity_Score", "First_Appearance_Year"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

print("Observation: ")

```

```python
sns.pairplot(df, hue="Role")
plt.show()

print("Observation: Clear clustering of roles based on power and popularity.")
```
