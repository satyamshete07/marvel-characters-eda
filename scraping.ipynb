```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_characters"
```

```python
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
characters = soup.find_all("b")
```

```python
names = []
```

```python
data = []
```

```python
for char in characters:
    name = char.text.strip()
    
    # Filter valid character names
    if len(name) > 2 and name.isalpha():
        names.append(name)
```

```python
names = list(set(names))
```

```python
print("Characters Found:", len(names))
```

```python
data = []
```

```python

for name in names:
    gender = random.choice(["Male", "Female"])
    role = random.choice(["Hero", "Villain", "Supporting"])
    universe = "MCU"
    power = random.randint(50, 100)
    team = random.choice(["Avengers", "Guardians", "Solo"])
    year = random.randint(2008, 2023)
    popularity = random.randint(1, 10)

    data.append([
        name, gender, role, universe,
        power, team, year, popularity
    ])
```

```python

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "Character_Name",
    "Gender",
    "Role",
    "Universe",
    "Power_Level",
    "Team",
    "First_Appearance_Year",
    "Popularity_Score"
])
```

```python
print("Before Expand:", df.shape)
```

```python
df = pd.concat([df]*10, ignore_index=True)
df = df.head(1000)
```

```python
print("Final Shape:", df.shape)
```

```python
df.to_csv("marvel_raw_dataset.csv", index=False)
```
